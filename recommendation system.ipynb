```python
import json
```

```python
movies_0_genres = json.loads(movies.iloc[0].genres)  ## list
movies_0_genres = [element['name'] for element in movies_0_genres ]
movies_0_genres 
```

```python
def convert(genres_string):
    genres_dict = json.loads(genres_string)  ## we can also use ast.literal_eval(string)
    genres = [element['name'] for element in genres_dict]
    return genres
    
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
print([movies['genres']])
```

```python
%%capture
cast_0_list = json.loads(movies.iloc[0].cast)
cast_0_list

## So It's a list of several cast members with their character_name, original_name and other details
## Actors and Actress play an important role when a user wants to watch a movie, 
## I think the character name has least or no importance 
```

```python
len(cast_0_list)

## There are huge number of cast members for a single movie, Hence we shall consider 
## Top 5 cast which have important role in the film
## The 'order' key shows the cast member in the heirarcy of importance in the film
```

```python
no_of_casts_considered = 5
cast = [element['name'] for element in cast_0_list if element['order'] in range(0,no_of_casts_considered)]
```

```python
cast
```

```python
## Lets write a function that applies the above method to movies['cast'] series
def convert_cast(series):
    cast_list = json.loads(series)
    
    top_5_cast = [element['name'] for element in cast_list if element['order'] in range(0,no_of_casts_considered)]
    return top_5_cast
    
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies['cast'] 
## We have now, top 5 roles of a film
```

```python
crew_0_list = movies['crew'][0]
crew_0 = json.loads(crew_0_list)
crew_0
## This column has all the persons behind the scenes like production designer, sound designer, editor,writer etc
```

```python
for element in crew_0:
        print(element['job'],end= " / ")
    
## There are lot of people in the crew, but director is the major factor 
```

```python
def convert_crew(series):
    list = json.loads(series)
    dir = [element['name'] for element in list if element['job'] == 'Director']
    return dir
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)

## Now we have the director of every film in the crew column
```

```python
movies.iloc[0].overview
## It's a string of several words. We shall convert into list of words
## Desired output ['In', 'the', '22nd', 'century', ...]
```

```python
def convert_overview(series):
    return series.split(' ')
```

```python
movies['overview'].apply(convert_overview)
```

```python
movies['overview'] = movies['overview'].apply(convert_overview)
```

```python
movies.head(2)
```

```python
#### There is problem with the words with space inbetween, like 'James Cameron'
#### The model thinks that 'James' and 'Cameron' are two different entities
#### But in reality it's the name of a single person
#### Hence, we need to remove space and make it into a single word

#### Desired output 'JamesCameron'
```

```python
def remove_space(series):
    return [element.replace(' ','') for element in series]
        
```

```python
movies['crew'].apply(remove_space)
```

```python
## We shall apply this to cast, overview, keywords, genres
movies['crew'] = movies['crew'].apply(remove_space)
movies['cast'] = movies['cast'].apply(remove_space)
movies['overview'] = movies['overview'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
```

```python
movies.iloc[0]
## we observe, "Science Fiction" -> "ScienceFiction" in genres
## "Sam Worthington" -> "SamWorthington"             in overview
## culture clash -> "cultureclash"                   in keywords
```

```python
## Now, we shall create tags for each movie which has a bunch of words combined from the 5 columns 
## genres, cast, crew, keywords, overview

movies['tags'] = movies['genres'] + movies['cast'] +  movies['crew'] +  movies['keywords'] +  movies['overview']
```

```python
movies['tags'][0]
## This a list of words, We need to convert it into a single string
```

```python
def convert_list_to_string(tags):
    str  = ' '.join(tags)
    return  str
```

```python
movies['tags'].apply(convert_list_to_string)

```

```python
movies['tags'] = movies['tags'].apply(convert_list_to_string)
```

```python
movies.iloc[0]
```

```python
## So we can rid of those 5 columns since we have all the words in the tags column
## creating a new_df with 'movie_id', 'title', 'tags' only 
new_df = movies[['movie_id', 'title', 'tags']]
```
