```python
## Lets write a function that applies the above method to movies['cast'] series
def convert_cast(series):
    cast_list = json.loads(series)
    
    top_5_cast = [element['name'] for element in cast_list if element['order'] in range(0,no_of_casts_considered)]
    return top_5_cast
    
```

# TEXT VECTORIZATION

```python
## Tags column is not numerical, it has bunch of words
## Hence, we convert this text into vectors and then find the similarity scores

## We shall consider top 5000 words by frequency from all the movie tags combined
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
## stop words will neglect words like 'is', 'the', 'and', 'or', 'a' etc
## which actually doesn't carry any meaning of a statement.
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
## the vectors object is a scipy spare matrix
```

```python
cv.get_feature_names_out()
## This will show the 5000 words which are selected as feature by the model
```

```python
## This will show the 5000 words which are selected as feature by the model
for i in cv.get_feature_names_out():
    print(i)
```

```python
## We observe few words like 
# accident
# accidental
# accidentally

# The all 3 have the same meaning but they are now treated as 3 different features
# By method of stem-ing we can convert all 3 into same word like 'accident'


# Similarly, we can observe actor, actors as 2 different words
# We can convert them into single word 'actor' using stem method
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
```

```python
## working of stem method
w1 = stemmer.stem('actors')
w2 = stemmer.stem('actor')
print(w1, w2)

## every english word is associated with a root word
## Here root word of both actors and actor is actor
```

```python
## apply this stem function on all the tags and then generate the 5000 words
def stemming(series):
    return [stemmer.stem(tag) for tag in series.split()]
```

```python
new_df['tags'].apply(stemming)
## We can observe changes...
## adventure -> adventur
## fantasy -> fantasi
## scienceFiction -> sciencefict

## We can also observe all the words are converted into lower case
```

```python
new_df['tags'] = new_df['tags'].apply(stemming)

## we also need to convert that list into string 
new_df['tags'] = new_df['tags'].apply(convert_list_to_string)
```

```python
## So, we again need to fit the tags into countVectorizer 
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
for i in cv.get_feature_names_out():
    print(i)

## We can see stem is applied on words as well as numbers
```

```python
## cosine distance is the most valid measure to calculate distance between two vectors 
## cosine similarity is the inverse of cosine distance
## higher the value of cosine-similarity between two vectors (say movies), higher is 
## the probability that they both are alike

from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
similarity.shape
## Every row has the similarity score of that movie with all the 4806 movies (including it's own)
## Hence it's a matrix of 4806 x 4806
```

```python
import numpy as np
```

```python
## For a given movie, we shall consider the top 5 movies that have the highest similarity score in the row
## Hence, these are the 5 recommended movies for that given particular movie
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    similarity_scores = similarity[movie_index]
    top_5_movie_index = np.argsort(similarity_scores)[-6:-1]
    recommended_movies = []
    
    for i in reversed(top_5_movie_index): 
        recommended_movies.append(new_df.iloc[i].title)
        print(new_df.iloc[i].title)

    return recommended_movies
```

```python
recommend('Avatar')
```
